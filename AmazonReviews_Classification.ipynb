```python
stopwords = nltk.corpus.stopwords.words('english')
```

```python
def saveReviewsByYear(filename,content):
    with open(filename,'w') as file:
        for item in content:
            file.write("%s\n" %item)
    file.close()
```

```python
def filterReviewTextsByYear(year):
    tlist=rawText.split('\n\n')
    filteredlist=[]
    for elem in tlist:
        t1 = elem.split('\n')
        if(t1[len(t1)-1].find(year)!=-1):
            if t1[4].find('reviewText')!=-1:
                t1[4]=t1[4].replace('reviewText:','')
                filteredlist.append(t1[4])
            elif t1[3].find('reviewText')!=-1:
                t1[3]=t1[3].replace('reviewText:','')
                filteredlist.append(t1[3])
    return filteredlist
```

```python
filteredReviewTexts = filterReviewTextsByYear('2013')
```

```python
filteredReviewTexts[:5]
```

```python
def tokenizeFilteredReviews():
    tokenized_List = []
    PATTERN = r'''(?x) [a-zA-z]+'[a-zA-z]+ | [a-zA-z]+
               '''
    tokenized_List=[nltk.regexp_tokenize(review_text,PATTERN) for review_text in filteredReviewTexts]
    return tokenized_List
```

```python
tokenizedReviewsList = tokenizeFilteredReviews() 
```

```python
tokenizedReviewsList[0]
```

```python
def filterNonAlphabets(w):
    pattern = re.compile('^[^a-z]+$')
    if (pattern.match(w)):
        return True
    else:
        return False
```

```python
def toLower(list):
    lowerCaseList=[]
    for word in list:
        lowerCaseList.append(word.lower())
    return lowerCaseList
```

```python
def filterNonAlphaFromList(List):
    nonAlphalist=[]
    for word in List:
        if (filterNonAlphabets(word) == False):
            nonAlphalist.append(word)
    return nonAlphalist
```

```python
def filterStopWordsFromList(List):
    listWOStopWords = []
    for word in List:
        if word not in stopwords:
            listWOStopWords.append(word)
    return listWOStopWords
```

```python
def preProcessing():
    preProcessedReviewTokens = []
    for innerReviewList in tokenizedReviewsList:
        lowerCaseList= toLower(innerReviewList)
        nonAlphaList= filterNonAlphaFromList(lowerCaseList);
        listWOStopWords= filterStopWordsFromList(nonAlphaList)           
        preProcessedReviewTokens.append(listWOStopWords)
    return preProcessedReviewTokens
```

```python
preProcessedReviewTokens = preProcessing()
```

```python
preProcessedReviewTokens[:10]
```

```python
sentences = [(sentence, catg) for catg in sentence_polarity.categories() for sentence in sentence_polarity.sents(categories=catg)]
```

```python
sentences[:10]
```

```python
def wordFeatures():
    random.shuffle(sentences)
    wordlist = [word for (sentence,catg) in sentences for word in sentence]
    stopWordsRemovedList=filterStopWordsFromList(wordlist)       
    finalwordlist=filterNonAlphaFromList(stopWordsRemovedList)
    finalwords = nltk.FreqDist(finalwordlist)
    wordItems = finalwords.most_common(2000)
    wordFeatures = [ word for (word,frequency) in wordItems]
    return wordFeatures
```

```python
wordFeatures = wordFeatures()
```

```python
wordFeatures[:100]
```

```python
def getFeaturesForNB(sentence):
    # check if the sentence's word is present in the word features of above. If yes, then list as tuples
    sentence_words = set(sentence)
    features = {}
    for word in wordFeatures:
        features['contains({})'.format(word)] = (word in sentence_words)
    return features
```

```python
def GetfeatureSetForNBSentences():
    featureSetFromSentences = [(getFeaturesForNB(sent), catg) for (sent,catg) in sentences]
    return featureSetFromSentences
```

```python
def GetfeatureSetForNBReviews():
    featureSetFromReviews = [getFeaturesForNB(reviewText) for reviewText in preProcessedReviewTokens]
    return featureSetFromReviews
```

```python
featureSet1FromSentences = GetfeatureSetForNBSentences()
```

```python
featureSet1FromReviews = GetfeatureSetForNBReviews()
```

```python
sentiments=[]
def NaiveBayesClassification():
    train_set = featureSet1FromSentences[1000:]
    test_set = featureSet1FromSentences[:1000]
    classifier = nltk.NaiveBayesClassifier.train(train_set)
    accuracy = nltk.classify.accuracy(classifier, test_set)
    classifier.show_most_informative_features(20)
    sentiments = [ classifier.classify(feature) for feature in featureSet1FromReviews]
    pos_sentences=[]
    neg_sentences=[]
    i=0
    for value in sentiments:
        if(value == 'neg'):
            neg_sentences.append(filteredReviewTexts[i])
        else:
            pos_sentences.append(filteredReviewTexts[i])
        i+=1   
        
    return pos_sentences,neg_sentences,accuracy
```

```python
def readSubjectivity():
    file = open('subjclueslen1-HLTEMNLP05.tff', 'r')
    # initialize an empty dictionary
    sdict = { }
    for line in file:
        data = line.split()
        strength = data[0].split("=")[1]
        word = data[2].split("=")[1]
        posTag = data[3].split("=")[1]
        stemmed = data[4].split("=")[1]
        polarity = data[5].split("=")[1]
        if (stemmed == 'y'):
            sdict[word] = [strength, posTag, True, polarity]
        else:
            sdict[word] = [strength, posTag, False, polarity]
    return sdict
```

```python
wordDictionary = readSubjectivity()
```

```python
def getFeaturesForNB2(sentence):
    uwords = set(sentence)
    features = {}
    for word in wordFeatures:
        features['contains(%s)' % word] = (word in uwords)
        strongNeg = 0
        weakNeg = 0
        weakPos = 0
        strongPos = 0 
        for word in uwords:
            if word in wordDictionary:
                strength, posTag, isStemmed, polarity = wordDictionary[word]
                if strength == 'weaksubj' and polarity == 'positive':
                    weakPos += 1
                if strength == 'strongsubj' and polarity == 'positive':
                    strongPos += 1
                if strength == 'weaksubj' and polarity == 'negative':
                    weakNeg += 1
                if strength == 'strongsubj' and polarity == 'negative':
                    strongNeg += 1
                features['positivecount'] = weakPos + (2 * strongPos)
                features['negativecount'] = weakNeg + (2 * strongNeg)      
    return features
```

```python
def GetfeatureSet2ForNBSentences():
    featureSetFromSentences = [(getFeaturesForNB2(sent), catg) for (sent,catg) in sentences]
    return featureSetFromSentences
```
